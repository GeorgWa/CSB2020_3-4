```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import matplotlib as mpl
from matplotlib_venn import venn2
from matplotlib.collections import PatchCollection
from matplotlib.patches import Rectangle

sgd = pd.read_csv("Reschke_yeast_filtered.tsv",sep='\t')


    
def corr(df1,df2):
    merge = pd.merge(df1, df2, how='inner', on="id")
    merge['worst_fdr'] = merge.apply(lambda row: np.min([row['fdr_x'], row['fdr_y']]), axis=1)
    merge = merge[(merge.worst_fdr < 0.05)]
    
    x = merge.fc_x.to_numpy()
    y = merge.fc_y.to_numpy()
    cor, p = stats.pearsonr(x,y)
    
    print(cor)
    plt.scatter(x,y, s=0.2)
    ax = plt.gca()
    plt.text(0.05, 0.95, "R={:.2f}".format(cor), transform=ax.transAxes,verticalalignment='top')
    fig = plt.gcf()
    fig.set_size_inches(8, 8)
    
```
